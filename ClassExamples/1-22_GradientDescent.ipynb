```python
import numpy as np

X_train = np.array([1, 2, 3, 4, 4, 5, 6])
Y_train = np.array([40, 80, 130, 140, 150, 170, 173])
N = 7
theta0 = 40
theta1 = 40
errors = []

for i in range(N):
    yhat = theta1 * X_train[i] + theta0
    e = (yhat - Y_train[i]) ** 2
    errors.append(e)

E = sum(errors) / N
print(E)
```
