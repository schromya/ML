```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
x_train = np.array([[120], [125], [161], [119], [118], [117], [159], [150]])
y_train = np.array([[0],  [0], [1], [0], [0], [0], [1], [1]])
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
healthy_index = y_train == 0
notHealthy_index = y_train == 1
plt.scatter(x_train[healthy_index], y_train[healthy_index], c ="green")
plt.scatter(x_train[notHealthy_index], y_train[notHealthy_index], c ="red")



xfit = np.linspace(70, 200, 100)
yfit = lr.predict(xfit[:, np.newaxis])
plt.plot(xfit, yfit);
myThresh=np.where(np.logical_and(yfit>=0.49, yfit<=0.51))
plt.axvline(x = xfit[myThresh[0]], color = 'k')
plt.show()
```

```python
import numpy as np 
def sig(t):
 return 1/(1 + np.exp(-t))
```

```python
t = 10
print(sig(t))
```

```python
t = 100
print(sig(t))
```

```python
t = -100
print(sig(t))
```

```python
t = 0
print(sig(t))
```

```python
t = np.linspace(-10, 10, 1000)   
mySig = sig(t)
plt.xlabel("t") 
plt.ylabel("Sigmoid(t)")  
plt.plot(t, mySig) 
plt.show()
```
