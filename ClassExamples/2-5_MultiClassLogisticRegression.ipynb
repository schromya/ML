```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import sklearn.metrics as metrics
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
print(digits.data[0])
plt.imshow(np.reshape(digits.data[0], (8,8)))
```

```python
plt.figure(figsize=(25,5))
for ii, (image, label) in enumerate(zip(digits.data[0:7],
                                           digits.target[0:7])):
    plt.subplot(1, 7, ii + 1)
    plt.imshow(np.reshape(image, (8,8)))
    plt.title('Training: %i\n' % label, fontsize = 18);
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(digits.data,  digits.target, test_size=0.2, random_state=0)
X_train.shape, X_test.shape
```

```python
X_train.mean(axis=0)
```

```python
X_train.std(axis=0)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train.mean(axis=0).round(2)
```

```python
X_train.std(axis=0)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
model.coef_[0].round(2)
```

```python
plt.imshow(model.coef_[0].reshape(8,8));
```

```python
plt.figure(figsize=(10,5))

for i in range(10):
    coef_plot = plt.subplot(2, 5, i + 1)

    plt.imshow(model.coef_[i].reshape(8,8));
```

```python
print(model.predict(X_test[0:35]))
print(y_test[0:35])
```

```python
print(model.predict(X_test[36:70]))
print(y_test[36:70])
```

```python
plt.imshow(np.reshape(X_test[56], (8,8)))
```

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
predictions = model.predict(X_test)

confusion_matrix = metrics.confusion_matrix(y_true=y_test,  y_pred = predictions, labels = model.classes_)
```

```python
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix)
cm_display.plot()
plt.show()
```

```python
from sklearn.metrics import f1_score
f1_score(y_test, predictions, average=None)
```

```python
from sklearn.metrics import f1_score
f1_score(y_test, predictions, average='macro')
```
