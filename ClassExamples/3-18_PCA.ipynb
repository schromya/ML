```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import make_blobs
from sklearn.decomposition import PCA

# Generate sample data
X, y = make_blobs(n_samples=100, centers=3, n_features=3, random_state=42)

# Visualize original data in 3D plot
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y)
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.set_title('Original Data')

# 1. Standardize the data
X_std = (X - np.mean(X, axis=0)) / np.std(X, axis=0)

# 2. Compute the covariance matrix
covariance_matrix = np.cov(X_std.T)

# 3. Calculate the eigenvalues and eigenvectors of the covariance matrix
eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)

# 4. Sort the eigenvectors
k = 2
idx = eigenvalues.argsort()[::-1][:k]
top_eigenvectors = eigenvectors[:, idx]

# 5. Project the data onto the new k-dimensional space
X_pca = np.dot(X_std, top_eigenvectors)

# Visualize
mean = np.mean(X, axis=0)
ax.quiver(mean[0], mean[1], mean[2], top_eigenvectors[0, 0], top_eigenvectors[1, 0], top_eigenvectors[2, 0], color='red', length=3, normalize=True)
ax.quiver(mean[0], mean[1], mean[2], top_eigenvectors[0, 1], top_eigenvectors[1, 1], top_eigenvectors[2, 1], color='blue', length=3, normalize=True)
fig, ax = plt.subplots(figsize=(8, 8))
ax.scatter(X_pca[:, 0], X_pca[:, 1], c=y)
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_title('PCA Projection')

plt.show()
```
